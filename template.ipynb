## Step 4: Hypothesis Testing

```python
# TODO
import pandas as pd
import numpy as np

# 1. Create the 'under_30' column
df['under_30'] = df['minutes'] < 30

# 2. Calculate Observed Statistic
obs = df.groupby('under_30')['rating'].mean().diff().iloc[-1]

# 3. Run Permutation
metas = []
for _ in range(1000):
    shuffled_ratings = df['rating'].sample(frac=1).reset_index(drop=True)
    dist = df.assign(shuffled=shuffled_ratings).groupby('under_30')['shuffled'].mean().diff().iloc[-1]
    metas.append(dist)

# 4. Calculate P-value
p_val = np.mean(np.array(metas) >= obs)
print(f"Observed Diff: {obs}, P-value: {p_val}")
```

## Step 6: Baseline Model

```python
# TODO
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Selecting two simple features for the baseline
X = df[['minutes', 'n_steps']].fillna(0) # Simple fill for baseline
y = df['rating'].fillna(df['rating'].mean()) # Simple fill for target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

baseline_model = LinearRegression()
baseline_model.fit(X_train, y_train)

preds = baseline_model.predict(X_test)
rmse = mean_squared_error(y_test, preds, squared=False)
print(f"Baseline RMSE: {rmse}")
```
